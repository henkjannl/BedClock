### Filtering the capacitive sensors

The file ```Signal from top sensor.txt``` contains three deliberate sensor pushes, and then 10000 events where incidental noise is recorded. The purpose is to find an algorithm that accepts the first twee signals, and ignores all other events.

````python
import numpy as np
import matplotlib.pyplot as plt

signal=np.asarray([int(s) for s in open('Signal from top sensor.txt', 'r').readlines()])

plt.figure(figsize=(12,5))
plt.title('Full recording of the signal')
plt.plot(signal)
plt.show()
````

````python
accepted=signal[30:65]
ignored=signal[720:755]

plt.figure(figsize=(12,3))

plt.subplot(1,2,1)
plt.title('To be accepted:')
plt.grid(b=True, which='major', axis='x')
plt.plot(accepted)

plt.subplot(1,2,2)
plt.title('To be ignored:')
plt.grid(b=True, which='major', axis='x')
plt.plot(ignored)

plt.show()
````

Spikes are typically one sample wide. A median filter is proposed to filter out incidental spikes. We'd like to check if we can afford to use a median filter with kernel size 3, or that we need to use a wider kernel. To do that, we first check how many events remain if we use a larger kernel size.

````python
import scipy.signal

filtered=scipy.signal.medfilt(signal, kernel_size=7)
   
accepted=filtered[30:65]
ignored=filtered[720:755]

plt.figure(figsize=(12,3))

plt.subplot(1,2,1)
plt.title('Accepted:')
plt.grid(b=True, which='major', axis='x')
plt.plot(accepted)

plt.subplot(1,2,2)
plt.title('Ignored:')
plt.grid(b=True, which='major', axis='x')
plt.plot(ignored)

plt.show()
````
