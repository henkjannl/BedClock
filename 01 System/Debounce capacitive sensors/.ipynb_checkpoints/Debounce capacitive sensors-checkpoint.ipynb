## Filtering the capacitive sensors

The file ```Signal from top sensor.txt``` contains:
* three long sensor presses
* three short sensor presses
* 10000 samples where incidental spikes are recorded 

The goal is to find an algorithm that accepts the first 5 deliberate pushes, and ignores all spikes.

````python
import numpy as np
import matplotlib.pyplot as plt

signal=np.asarray([int(s) for s in open('Signal from top sensor.txt', 'r').readlines()])

plt.figure(figsize=(12,5))
plt.title('Full recording of the signal')
plt.plot(signal)
plt.show()
````

````python
accepted     = signal[ 30: 60]
alsoAccepted = signal[140:170]
ignored      = signal[720:755]

plt.figure(figsize=(12,3))

plt.subplot(1,3,1)
plt.title('Short press:')
plt.grid(b=True, which='major', axis='x')
plt.plot(accepted)

plt.subplot(1,3,2)
plt.title('Long press:')
plt.grid(b=True, which='major', axis='x')
plt.plot(alsoAccepted)

plt.subplot(1,3,3)
plt.title('Sensor noise:')
plt.grid(b=True, which='major', axis='x')
plt.plot(ignored)

plt.show()
````

## Applying a median filter
Sensor noise is typically one sample wide. A median filter is proposed to filter incidental spikes. We'd like to check if we can afford to use a median filter with kernel size 3, or that we must use a wider kernel. To do that, we first check how many events remain if we use a larger kernel size.

````python
import scipy.signal

plt.figure(figsize=(12,12))

graphRow=0
for kernel in [7,5,3]:
    filtered=scipy.signal.medfilt(signal, kernel_size=kernel)

    accepted=filtered[30:65]
    alsoAccepted=filtered[140:165]
    ignored=filtered[720:755]

    plt.subplot(3,3,3*graphRow+1)
    plt.title('Kernelsize %d, long press:' % kernel)
    plt.grid(b=True, which='major', axis='x')
    plt.plot(accepted)

    plt.subplot(3,3,3*graphRow+2)
    plt.title('Kernelsize %d, short press:' % kernel)
    plt.grid(b=True, which='major', axis='x')
    plt.plot(alsoAccepted)

    plt.subplot(3,3,3*graphRow+3)
    plt.title('Kernelsize %d, noise:' % kernel)
    plt.grid(b=True, which='major', axis='x')
    plt.plot(ignored)
    
    graphRow+=1

plt.show()
````

````python
filtered=scipy.signal.medfilt(signal, kernel_size=3)

plt.figure(figsize=(12,3))
plt.title('Full range')
plt.plot(filtered<65)
plt.show()

plt.figure(figsize=(12,3))
plt.title('First 300 samples')
plt.plot(filtered[0:300]<65)
plt.show()
````

**Conclusions:** 
* kernel sizes of 5 and 7 are effective to filter out noise, but will also filter out the short presses
* kernal size 3 effectively filters out noise and retains the short presses
* we may want to increase the scanning frequency and go for kernel size 5

## Checking if it works
An efficient median filter was used from Travis Mick's implemention: 

https://github.com/tmick0/generic_median

The filter implements the Hoare's in-place selection algorithm to obtain the median. Since the implementation of the algorithm returns gibberish the first number of samples, the filter is loaded with a couple raw sensor values upon initialization.

![Sir Charles Antony Richard Hoare](https://upload.wikimedia.org/wikipedia/commons/thumb/2/2c/Sir_Tony_Hoare_IMG_5125.jpg/240px-Sir_Tony_Hoare_IMG_5125.jpg)
<center>Sir Charles Anthony Richard Hoare</center>

The log file below reports the raw sensor signal, plus the filtered signal using a median filter with kernel size 3.


````python
import numpy as np
import matplotlib.pyplot as plt

lines=open('Result from median kernel 3.txt', 'r').readlines()
raw    = np.asarray([ int(s.split('\t')[0]) for s in lines ])
med = np.asarray([ int(s.split('\t')[1]) for s in lines ])

plt.figure(figsize=(12,3))
plt.title('Sensor signal, full range')
plt.plot(raw)
plt.show()

plt.figure(figsize=(12,3))
plt.title('Filtered signal, full range')
plt.plot(med)
plt.show()

plt.figure(figsize=(12,3))
plt.title('First 300 samples')
plt.grid(b=True, which='major', axis='y')
plt.plot(med[0:300])
plt.show()

plt.figure(figsize=(12,3))
plt.title('Threshold of 65 applied')
plt.plot(med[0:300]<65)
plt.show()
````
